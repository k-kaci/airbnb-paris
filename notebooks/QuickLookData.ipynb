- Il faut mettre les prix au bon format en enlevant le $

```python
def convert_prices(price):
    if pd.isna(price):
        return price
    price = np.na = price.split('$')[1]
    price_split = price.split(',')
    if(len(price_split) == 2):
        price = price_split[0] + price_split[1]
    return pd.to_numeric(price)
```

```python
listings_selected['price'] = listings_selected['price'].map(convert_prices)
```
